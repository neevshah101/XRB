```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import pandas as pd
plt.style.use(['science','grid','bright'])
%matplotlib inline
```

```python
logs_dir_B = '/home/u16/neevshah/Projects/XRB/binary/m1_60_m2_40_p_6_B_NL_D'
donor_B = pd.read_table(logs_dir_B+'/LOGS1/history.data', sep = '\s+', skiprows = 5)
binary_B = pd.read_table(logs_dir_B+'/binary_history.data', sep = '\s+', skiprows = 5)

logs_dir_S = '/home/u16/neevshah/Projects/XRB/binary/m1_60_m2_40_p_6_S_NL_D'
donor_S = pd.read_table(logs_dir_S+'/LOGS1/history.data', sep = '\s+', skiprows = 5)
binary_S = pd.read_table(logs_dir_S+'/binary_history.data', sep = '\s+', skiprows = 5)
```

```python
plt.plot(binary_B['age'], 10**donor_B['log_Teff'])
plt.plot(binary_S['age'], 10**donor_S['log_Teff'])
```

```python
lw = 2
plt.plot(10**donor_B['log_Teff'], donor_B['log_L'], label = 'Bjorklund', lw = lw)
plt.plot(10**donor_S['log_Teff'], donor_S['log_L'], label = 'Sabhahit', lw = lw)
plt.gca().invert_xaxis()
plt.legend()
plt.show()

lw = 2
plt.plot(donor_B['star_age'], binary_B['lg_wind_mdot_1'], label = 'Bjorklund', lw = lw)
plt.plot(donor_S['star_age'], binary_S['lg_wind_mdot_1'], label = 'Sabhahit', lw = lw)
plt.legend()
plt.show()

lw = 2
plt.plot(donor_B['star_age'], binary_B['star_1_radius'], lw = lw, color = 'C0')
plt.plot(donor_B['star_age'], binary_B['rl_1'], lw = lw, color = 'C0', ls = '--')
plt.plot(donor_S['star_age'], binary_S['star_1_radius'], lw = lw, color = 'C1')
plt.plot(donor_S['star_age'], binary_S['rl_1'], lw = lw, color = 'C1', ls = '--')
plt.show()
```
