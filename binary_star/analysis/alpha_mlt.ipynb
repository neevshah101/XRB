```python
# Exclude rows from position 26000 to 60000
filtered_df = pd.concat([df[:26000], df[60000:]])
other_df = df[26000:60000]
```

```python
# Exclude rows from position 26000 to 60000
filtered_dfh = pd.concat([dfh[:26000], dfh[60000:]])
other_dfh = dfh[26000:60000]
```

```python
import matplotlib.pyplot as plt

lw = 2

# Create a figure with 2 rows and 1 column of subplots
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(4, 12), sharex=True)

# First plot: Mass vs Age
axes[0].plot(filtered_df['age'] / 1e6, filtered_df['star_1_mass'], lw=lw, label='Star 1')
axes[0].plot(filtered_df['age'] / 1e6, filtered_df['star_2_mass'], lw=lw, label='Star 2')
axes[0].legend()
axes[0].set_ylabel('Mass (Msun)')
axes[0].set_xlim(3, 4.6)

# Second plot: Relative Overflow vs Age
axes[1].plot(filtered_df['age'] / 1e6, filtered_df['rl_relative_overflow_1'], lw=lw, label='RLOF')
axes[1].legend()
axes[1].set_xlabel('Age (Myr)')
axes[1].set_ylabel('Relative Overflow')
axes[1].set_xlim(3, 4.6)

# Second plot: Relative Overflow vs Age
axes[2].plot(filtered_df['age'] / 1e6, filtered_df['rl_1'], lw=lw, label='RL1')
axes[2].plot(filtered_df['age'] / 1e6, filtered_df['binary_separation'], lw=lw, label='Separation')
axes[2].plot(filtered_df['age'] / 1e6, filtered_df['star_1_radius'], lw=lw, label='Star 1')
axes[2].legend()
axes[2].set_xlabel('Age (Myr)')
axes[2].set_ylabel('Rsun')
axes[2].set_xlim(3,6)

# Second plot: Relative Overflow vs Age
axes[3].plot(filtered_df['age'] / 1e6, filtered_df['lg_mtransfer_rate'], lw = lw, label = 'MT')
axes[3].plot(filtered_df['age'] / 1e6, filtered_df['lg_wind_mdot_1'], lw=lw, label='Wind 1')
axes[3].legend()
axes[3].set_xlabel('Age (Myr)')
axes[3].set_ylabel('Msun/yr')
axes[3].set_ylim(-9,-2)
axes[3].set_xlim(3, 4.6)

# Adjust layout
plt.tight_layout()
plt.show()
```
